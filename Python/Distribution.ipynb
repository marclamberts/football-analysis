```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file
file_path = 'NLXG.xlsx'  # Replace with your actual file path
sheet_name = 'Sheet1'  # Replace with your actual sheet name, if needed
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Assuming you have a column named 'xG' which you want to plot the distribution for
xg_column = 'xG'  # Replace with your actual xG column name in the Excel file
xg_values = data[xg_column]

# Set a fancier style (you can try 'ggplot', 'seaborn-darkgrid', 'fivethirtyeight', etc.)
plt.style.use('seaborn-darkgrid')

# Plot the distribution (histogram) with more enhancements
plt.figure(figsize=(10, 6))
plt.hist(xg_values, bins=30, edgecolor='white', alpha=0.85, color='skyblue')

# Adding mean and median lines
mean_xg = np.mean(xg_values)
median_xg = np.median(xg_values)

plt.axvline(mean_xg, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_xg:.2f}')
plt.axvline(median_xg, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median_xg:.2f}')

# Add titles and labels with bigger fonts
plt.title('Distribution of xG (Expected Goals)', fontsize=16, fontweight='bold')
plt.xlabel('xG (Expected Goals)', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Adding a legend
plt.legend()

# Improve grid visibility
plt.grid(True, alpha=0.6)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file
file_path = 'NLXG.xlsx'  # Replace with your actual file path
sheet_name = 'Sheet1'  # Replace with your actual sheet name, if needed
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Assuming you have columns named 'xG' and 'Minutes'
xg_column = 'xG'  # Replace with your actual xG column name in the Excel file
minutes_column = 'timeMin'  # Replace with your actual Minutes column name

# Extracting the xG and Minutes data
xg_values = data[xg_column]
minutes_values = data[minutes_column]

# Set the style for a fancier look
plt.style.use('seaborn-darkgrid')

# Plot the distribution (histogram)
plt.figure(figsize=(10, 6))

# Use a histogram where the x-axis is the 'Minutes' and aggregate the xG values into bins
plt.hist(minutes_values, bins=30, weights=xg_values, edgecolor='white', alpha=0.85, color='skyblue')

# Adding titles and labels
plt.title('Distribution of xG over Minutes', fontsize=16, fontweight='bold')
plt.xlabel('Minutes', fontsize=14)
plt.ylabel('xG (Expected Goals)', fontsize=14)

# Adding a grid for better readability
plt.grid(True, alpha=0.6)

# Show the plot
plt.show()
```
