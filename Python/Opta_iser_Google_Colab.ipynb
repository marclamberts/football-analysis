```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Sample DataFrame (replace this with your actual data)
# df = pd.read_csv('your_data.csv') # Uncomment if you're loading from a CSV

# Add calculated features
df["distance_to_goal"] = np.sqrt((105 - df["endX"])**2 + (34 - df["endY"])**2)
df["angle_to_goal"] = np.arctan2(34 - df["endY"], 105 - df["endX"])

# Encode targets based on rules
df["goal"] = (df["typeId"] == 16).astype(int)
df["loss"] = (df["typeId"] == 9).astype(int)
df["possession_continuation"] = ((df["typeId"] == 1) & (df["outcome"] == 1)).astype(int)

# Add new encoded columns
df["TakeO"] = (df["typeId"] == 3).astype(int)  # TypeId == 3 for TakeOn
df["Shot"] = ((df["typeId"] == 13) | (df["typeId"] == 14) | (df["typeId"] == 15)).astype(int)  # TypeId 13, 14, or 15 for Shot
df["Tackle"] = ((df["typeId"] == 7) & (df["outcome"] == 1)).astype(int)  # TypeId == 7 and Outcome == 1 for Tackle
df["Ball_recovery"] = (df["typeId"] == 49).astype(int)  # TypeId == 49 for Ball Recovery
df["Assists"] = (df["Assist"] == 1).astype(int)  # Assists column equal to 1
df["2nd_Assists"] = (df["2nd Assist"] == 1).astype(int)  # 2nd Assist column equal to 1

# Features and targets
features = ["x", "y", "endX", "endY", "distance_to_goal", "angle_to_goal",
            "TakeO", "Shot", "Tackle", "Ball_recovery", "Assists", "2nd_Assists"]
targets = ["goal", "possession_continuation", "loss", "TakeO", "Shot", "Tackle", "Ball_recovery", "Assists", "2nd_Assists"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    df[features], df[targets], test_size=0.2, random_state=42
)

# Train models for each outcome
models = {}
for target in targets:
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train[target])
    models[target] = model

# Initialize arrays to accumulate probabilities
probs = {target: np.zeros(len(X_test)) for target in targets}

# Run 1000 simulations
n_simulations = 1
for _ in range(n_simulations):
    for target in targets:
        if len(models[target].classes_) == 1:
            probs[target] += 0  # Handle single-class case
        else:
            probs[target] += models[target].predict_proba(X_test[features])[:, 1]

# Average probabilities over simulations
for target in targets:
    probs[target] /= n_simulations

# Add probabilities to X_test
for target in targets:
    X_test[f"{target}_prob"] = probs[target]

# Define weights for each metric's contribution to EPV
weights = {
    "goal_prob": 1.5,                # Goals have the highest impact
    "loss_prob": -1.0,              # Losses negatively impact EPV
    "possession_continuation_prob": 0.8,  # Continuation keeps possession
    "TakeO_prob": 0.5,              # Take-ons slightly increase possession value
    "Shot_prob": 1.2,               # Shots are key offensive actions
    "Tackle_prob": 0.4,             # Tackles regain possession
    "Ball_recovery_prob": 0.6,      # Ball recoveries regain possession
    "Assists_prob": 1.0,            # Assists directly lead to goals
    "2nd_Assists_prob": 0.7         # 2nd assists indirectly contribute
}

# Calculate a weighted EPV using the probabilities
X_test["epv"] = sum(
    weights[f"{metric}_prob"] * X_test[f"{metric}_prob"]
    for metric in targets
)

# Display the first few rows to verify
print(X_test[[f"{metric}_prob" for metric in targets] + ["epv"]].head())
```
