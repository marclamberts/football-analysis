```python
import pandas as pd
import seaborn as sns
from IPython.display import display

# Load the Excel file into a dataframe
#df = pd.read_excel('/Users/marclambertes/Downloads/Scouting men 2324.xlsx')
df = pd.read_excel('/Users/marclambertes/Downloads/Total/Eredivisie - 2023-2024.xlsx')

# Filter the dataframe based on criteria
df = df[df["Minutes played"] >= 900]
#df = df[df["Age"] <= 27]
df = df[df["Position"].isin(["CF"])]

# Sort the dataframe by "Goals per 90" column in descending order
sorted_df = df.sort_values("xG per 90", ascending=False)

# Select top 50 players
top_50_players = sorted_df.head(50)

# Select desired columns for display
columns_to_show = ["Player", "Team", "Team within selected timeframe", "Position", "Age", "Minutes played", 
                   "Goals per 90", "xG per 90", "Shots per 90", "Assists per 90", "xA per 90", "Dribbles per 90", "Touches in box per 90"]
top_50_players_subset = top_50_players[columns_to_show]

# Set the color gradient for "Goals per 90" column
cm = sns.light_palette("green", as_cmap=True)
top_50_players_styled = top_50_players_subset.style.background_gradient(cmap=cm)

# Format numeric columns with two decimal places
numeric_cols = ["Goals per 90", "xG per 90", "Shots per 90", "Assists per 90", "xA per 90", "Dribbles per 90", "Touches in box per 90"]
format_dict = {col: "{:.2f}" for col in numeric_cols}
format_dict["Age"] = "{:.0f}"
top_50_players_styled = top_50_players_styled.format(format_dict)

# Save the styled dataframe to an Excel file
top_50_players_styled.to_excel("Strikers Eerste Divisie .xlsx", index=False)

top_50_players_styled
```
