```python
import pandas as pd
from scipy.stats import hmean

# Load the Excel file into a DataFrame
file_path = "/Users/marclambertes/Downloads/Database Men 2023-2024/Croatia - 2023-2024.xlsx"
df = pd.read_excel(file_path)

# Drop rows with NaN values in the specified columns
df.dropna(subset=['Smart passes per 90', 'Deep completions per 90', 'Key passes per 90', 'Shot assists per 90'], inplace=True)

# Filter out rows where any of the metrics are zero
df = df[(df['Smart passes per 90'] != 0) & 
        (df['Deep completions per 90'] != 0) & 
        (df['Key passes per 90'] != 0) & 
        (df['Shot assists per 90'] != 0)]

# Calculate harmonic mean for each player
df['Harmonic Mean'] = hmean(df[['Smart passes per 90', 'Deep completions per 90', 'Key passes per 90', 'Shot assists per 90']], axis=1)

# Print player names along with their specific harmonic mean
for index, row in df.iterrows():
    print(f"Player: {row['Player']}, Harmonic Mean: {row['Harmonic Mean']}")
    
df.to_excel("/Users/marclambertes/Downloads/PDI Eredivisie.xlsx", index=False)

```
